## Предобработка данных

Посчитаем количество пропусков и их долю

```python
def nan_values(data):
    count_missing = data.isna().sum()
    percent_missing = round(data.isna().sum() * 100 / len(data), 2)
    
    missing_value_data = pd.DataFrame({'count_missing': count_missing,
                                     'percent_missing': percent_missing})
    
    return missing_value_data

nan_values(data)
```

###  Заполнение пропусков

#### Высота потолков 'ceiling_height'
В столбце 'ceiling_height' заменим пропуски медианным значением высоты потолков в зависимости от населённого пункта и количества этажей в доме. 

```python
data['ceiling_height'].fillna(
    data.groupby(['locality_name', 'floors_total'])
    ['ceiling_height'].transform('median'), inplace=True
)
```

```python
data['ceiling_height'].isna().sum()
```

Остаётся 503 пропуска высоты потолков, заменим их медианой по столбцу.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

#### Апартаменты 'is_apartment'
В столбце 'is_apartment' заменим пропуски на False.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

#### Балконы 'balcony'
В столбце 'balcony' пропуски означают что балкона нет, поэтому заменим на 0.

```python
data['balcony'] = data['balcony'].fillna(0)
```

#### Площадь кухни 'kitchen_area' и жилая площадь 'living_area'
Для замены пропусков в столце 'kitchen_area' найдём среднее от отношения площади кухни к общей площади.
Аналагично заменим пропуски в столбце 'living_area'. 

```python
data['kitchen_area'].fillna(
    (data['kitchen_area'] / data['total_area']).mean()
    * data['total_area'], inplace=True)

data['living_area'].fillna(
    (data['living_area'] / data['total_area']).mean() 
    * data['total_area'], inplace=True)
```

#### Число парков и прудов в радиусе 3 км и расстояние до ближайшего парка и пруда

В столбце 'parks_around3000'(количество парков в радиусе 3км) пустые значения скорее всего означают что в радиусе 3км парков нет, по этим пропускам в столбце 'parks_nearest'(расстояние до ближайшего парка) данных тоже нет. Заменим пропуски на 0.
А в столбце 'parks_nearest' оставим отсутствующие значения т.к. для их заполнения у нас нет данных и в нашем исследовании этот параметр не используется.
Аналогичная ситуация со столбцами 'ponds_around3000'(количество прудов в радиусе 3км) и 'ponds_nearest'(расстояние до ближайшего пруда).

```python
data['parks_around3000'] = data['parks_around3000'].fillna(0)
```

```python
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
```

####  Cколько дней было размещено объявление 'days_exposition'
Пропуски в 'days_exposition' означают что объявление еще не снято с публикации.
В этих случаях правильно оставить эти значения пустыми, чтобы не исказить данные.

####  'cityCenters_nearest' , 'airports_nearest', 'floors_total', 'locality_name'
Расстояние до центра города и расстояние до аэропорта заполнить нет возможности, чтобы найти это расстояние нам нужен например адресс квартиры. 
Пропуски в столбце 'floors_total' составляют 0.36% данных и столбце 'locality_name'составляют 0.21%. Если мы удалим строки с пропусками это не повлияют не результат исследования.

```python
data = data.dropna(subset=['locality_name', 'floors_total'])
```

## Расчёты и добавление результатов в таблицу

Посчитаем и добавим в таблицу: цена квадратного метра.

```python
data['price_metre'] = round(data['last_price']/data['total_area'])
```

Посчитаем и добавим в таблицу: день недели, месяц и год публикации объявления

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'])
```

```python
data['day_of_week'] = data['first_day_exposition'].dt.dayofweek
data['month'] = data['first_day_exposition'].dt.month
data['year'] = data['first_day_exposition'].dt.year
```

```python
data['month_year'] = data['first_day_exposition'].apply(lambda x:x.strftime('%B-%Y'))
```

Посчитаем и добавим в таблицу: этаж квартиры; варианты — первый, последний, другой.

```python
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'Первый'
    if floor >= floors_total:
        return 'Последний'
    return 'Другой'
```

```python
data['floor_category'] = data.apply(floor_category, axis=1)
```

Посчитаем и добавим в таблицу: соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
data['living_area_ratio'] = data['living_area'] / data['total_area'] * 100
```

```python
data['kitchen_area_ratio'] = data['kitchen_area'] / data['total_area'] * 100
```
